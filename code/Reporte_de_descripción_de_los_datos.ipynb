| Variable | Tipo Inicial | Significado |
|----------|--------------|-------------|
| `video_id` | `object` | Identificador único alfanumérico del video. |
| `trending_date` | `object` | Fecha en la que el video alcanzó tendencia. |
| `title` | `object` | Título del video. |
| `channel_title` | `object` | Nombre del canal que publicó el video. |
| `category_id` | `float64` | Código numérico asociado a la categoría del video. |
| `publish_time` | `object` | Fecha y hora exacta de publicación del video. |
| `tags` | `object` | Etiquetas o palabras clave asociadas al video. |
| `views` | `float64` | Conteo total de visualizaciones acumuladas. |
| `likes` | `float64` | Cantidad total de "me gusta" recibidos. |
| `dislikes` | `float64` | Cantidad total de "no me gusta" recibidos. |
| `comment_count` | `float64` | Conteo total de comentarios en el video. |
| `thumbnail_link` | `object` | URL de la imagen en miniatura del video. |
| `comments_disabled` | `object` | Indicador booleano de si los comentarios están desactivados. |
| `ratings_disabled` | `object` | Indicador booleano de si las valoraciones están desactivadas. |
| `video_error_or_removed` | `object` | Indicador booleano de si el video fue eliminado o presenta error. |
| `description` | `object` | Descripción detallada proporcionada por el autor del video. |
| `state` | `object` | Estado geográfico asociado a la tendencia del video. |
| `lat` | `float64` | Latitud de la ubicación geográfica del registro. |
| `lon` | `float64` | Longitud de la ubicación geográfica del registro. |
| `geometry` | `object` | Representación geométrica (tipo `POINT`) de la ubicación. |
