## 1. Inventario de Datos

El presente proyecto utiliza el conjunto de datos "Trending YouTube Video Statistics". Los archivos consisten en registros diarios de videos en tendencia para el país asignado, México. Se han integrado archivos de metadatos en formato JSON para la descripción de categorías. Adicionalmente, el dataset ha sido enriquecido con coordenadas geográficas `(lat, lon)` y división administrativa `(state)`.

| Archivo | Formato | Peso Aprox.
| :--- | :--- | :---
| `MXvideos_cc50_202101.csv` | CSV | 45.7 MB |
| `MX_category_id.json` | JSON | 9 KB |

## 2. Metodología de Carga

La ingesta de los activos de datos se realizó mediante scripts en Python. Se empleó la librería Pandas para la carga del archivo CSV en una estructura de DataFrame, y el módulo estándar json para la lectura del archivo de metadatos. Se verificó la integridad de ambos archivos, confirmando que no existen errores de sintaxis en el JSON ni problemas de delimitadores en el CSV, permitiendo una carga fluida y sin incidentes técnicos.

```python
import pandas as pd
import numpy as np
import json

# 1. Carga del dataset transaccional
df = pd.read_csv('MXvideos_cc50_202101.csv')

# 2. Carga del archivo de categorías (metadatos)
with open('MX_category_id.json', 'r') as f:
    categories = json.load(f)

# 3. Verificación de la carga de los datos

resumen_datos = pd.DataFrame({
    "Variable": ["Dataset Transaccional", "Categorías (JSON)"],
    "Dimensiones / Ítems": [f"{df.shape[0]} filas, {df.shape[1]} columnas", f"{len(categories)} ítems"],
    "Fuente": ["MXvideos_cc50_202101.csv", "MX_category_id.json"]
})

from IPython.display import display
display(resumen_datos.style.set_caption("Tabla 1: Resumen de activos de datos recolectados"))
```
